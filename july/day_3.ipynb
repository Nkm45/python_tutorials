### Variable Assignments

#### Multiple Variable Assignment

Python also allows **Multiple Variable Assignment**, which essentially means more than 1 variable OR identifier (floating pointer) can be assigned to multiple literals.

```python

xa, mes3 = "XA", "MES3"

```

This means, `xa = "XA"` and `mes3 = "MES3"` in simple terms.

**Note:** Ensure the number of variables on LHS is equal to number of literals/objects on RHS. (Currently we're assuming this for variable assignments only)



### NOTE

When we create an object and is being used by a program (python script), we can create or delete multiple references/variable assignments to it as per our requirements. When the number of references (pointers pointing) to a particular object becomes **zero**, it is **no longer accessible**!

Python keeps a track of unused/un-referred objects (**Literals**). And when it feels that the user is not using it, it deletes it so that the occupied memory could be released and could be used for something else. This process is called **"Garbage Collection"** in CS.

#### Smart Tip

For the sake of optimization, Python assigns lower memory addresses to values in range of **-5** to **256**.

### Variable Deletion

Since we saw how to create and address a variable, we'll focus on deleting a variable.

Usually deleting a variable is easy. Prefix `del` to a variable name and it gets deleted. Ensure that you don't prefix it while writing an expression. Otherwise, it'll throw **SyntaxError**.


Other ways of deleting variable references (we'll see these in more detail later):
- Setting value as None (Not really)
- Using class methods (Only for iterables)

### Types of Variables

Generally speaking there are 2 types of variables, **Global** vars and **Local** vars and they operate exactly how they sound.

#### Global Variables

These are referred as **Public** variables in JAVA.

- Global variables are the ones which are available globally across your entire python script/file.
- They're accessible in your entire python script from any point.

#### Local Variables

These are referred as **Private** variables in JAVA.

- Opposite of Global variable.
- Accessible only in callable object (Classes OR Functions)
